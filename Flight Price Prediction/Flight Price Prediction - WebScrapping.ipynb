```python
airlinename = []
departuretime = []
arrivaltime = []
departurecity = []
arrivalcity = []
journeytime = []
numberofstops = []
flightprice = []


flightname = driver.find_elements_by_xpath("//div[@class='_3H-S _1Eia']")
flightname

for name in flightname:
    airlinename.append(name.text)

deptime = driver.find_elements_by_xpath("//div[@class='_3H-S']")
deptime

for dtime in deptime:
    departuretime.append(dtime.text)

arrtime = driver.find_elements_by_xpath("//div[@class='_16Wq _1OV0']/div[1]")
arrtime

for atime in arrtime:
    arrivaltime.append(atime.text.split("\n")[0])
    
depcity = driver.find_elements_by_xpath("//div[@class='_3Lds _1OV0']/div[2]")
depcity

for dcity in depcity:
    departurecity.append(dcity.text)

departurecity


arrcity = driver.find_elements_by_xpath("//div[@class='_16Wq _1OV0']/div[1]")
arrcity

for acity in arrcity:
    arrivalcity.append(acity.text.split("\n")[1])

    
joutime = driver.find_elements_by_xpath("//div[@class='vY4t']")
joutime

for jtime in joutime:
    journeytime.append(jtime.text)


noofstop = driver.find_elements_by_xpath("//div[@class='_7BOG']")
noofstop

for nstop in noofstop:
    numberofstops.append(nstop.text)

flyprice = driver.find_elements_by_xpath("//div[@class='_1cxG']")
flyprice

for fprice in flyprice:
    flightprice.append(fprice.text)

flightprice


flightpricedetails = pd.DataFrame()
flightpricedetails['name'] = airlinename
flightpricedetails['departue_time'] = departuretime
flightpricedetails['arrival_time'] = arrivaltime
flightpricedetails['departure_city'] = departurecity
flightpricedetails['arrival_city'] = arrivalcity
flightpricedetails['journey_time'] = journeytime
flightpricedetails['number_of_stops'] = numberofstops
flightpricedetails['price'] = flightprice
flightpricedetails

driver.close()
```
